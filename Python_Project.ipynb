4. Data Cleaning/Data Transformation

```python
import pandas as pd
data = [
    [105, 130, 0.02, 'none', 3.0, 'low', None, 'normal'],
    [106, 140, '?', 0.05, 'five', 0, 0, 1],
    [None, 120, 0.04, 0.02, 2, 'low', 'Zero', 'suspect'],
    [108, 'error', 0.07, 0.03, 1, None, 0, 'pathological'],
    [109, 128, 0.06, '', 2, 0, 0, 2],
    [110, 130, '-', '-', 'missing', 0, 0, 'suspect'],
    [111, 130, 0.01, 0.04, 2.0, 'low', 0, 'normal'],
    [None, 135, 0.05, 0.02, 2, 0, 1, 1],
    [113, 140, 0.03, 'nan', 1, 'low', 'Null', 2],
    [114, 140.0, 0.04, 0.01, 'nan', 0, 0, 'normal']
]
cols = ['patient_id', 'base_val', 'accels', 'movt', 'contractions', 'light_decels', 'sev_decels', 'fetal_health']

df = pd.DataFrame(data, columns=cols)
```

```python
df
```

```python
for col in ['base_val', 'accels', 'movt', 'contractions', 'sev_decels']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df['fetal_health'] = df['fetal_health'].replace({'normal':1, 'suspect':2, 'pathological':3})
df['fetal_health'] = pd.to_numeric(df['fetal_health'], errors='coerce')
```

```python
pd.set_option('future.no_silent_downcasting', True)
```

```python
df['light_decels'] = df['light_decels'].replace(
    {'low': 0, 'none': 0, 'null': 0, None: 0, 0: 0}
)
df['light_decels'] = df['light_decels'].infer_objects()
```

```python
cols_to_clean = ['base_val', 'accels', 'movt', 'contractions', 'sev_decels']
for col in cols_to_clean:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.fillna({
    'patient_id': -1,
    'base_val': df['base_val'].mean(),
    'accels': df['accels'].mean(),
    'movt': 0,
    'contractions': df['contractions'].median(),
    'sev_decels': 0,
    'light_decels': 0,
    'fetal_health': df['fetal_health'].mode()[0]
}, inplace=True)
```

```python
df['is_high_risk'] = (df['fetal_health'] == 3).astype(int)
```

```python
import pandas as pd
import numpy as np

# Sample data with outliers
data = {
    'patient_id': [201, 202, 203, 204, 205, 206, 207, 208, 209, 210],
    'base_val': [130, 132, 140, 128, 138, 400, 125, 127, 135, 133],       # 400 is an outlier
    'accels': [0.03, 0.04, 0.05, 0.045, 0.06, 0.021, 2.0, 0.05, 0.04, 0.03], # 2.0 is an outlier
    'movt': [0.02, 0.04, 0.01, 0.05, 0.04, 0.03, 0.02, 0.018, 0.017, 0.021]
}
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
df[['base_val', 'accels']].boxplot()
plt.title('Boxplot for Outlier Detection')
plt.show()
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_lim = Q1 - 1.5 * IQR
    upper_lim = Q3 + 1.5 * IQR
    # Filter out rows where column value is outside the bounds
    return df[(df[column] >= lower_lim) & (df[column] <= upper_lim)]

# For demonstration, treat outliers in 'base_val' and 'accels'
df_clean = remove_outliers_iqr(df, 'base_val')
df_clean = remove_outliers_iqr(df_clean, 'accels')

print("Before cleaning:", df.shape)
print("After cleaning:", df_clean.shape)
print(df_clean)
```

```python
def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_lim = Q1 - 1.5 * IQR
    upper_lim = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_lim, lower_lim, df[column])
    df[column] = np.where(df[column] > upper_lim, upper_lim, df[column])
    return df

# Cap outliers instead of removing
df_capped = cap_outliers_iqr(df.copy(), 'base_val')
df_capped = cap_outliers_iqr(df_capped, 'accels')
print(df_capped)
```
